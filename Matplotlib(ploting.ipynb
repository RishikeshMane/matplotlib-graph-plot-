### Using random data for ploting 

## Bar-plot

```python
x=[10,20,30,40,50]
y=[230,800,399,770,760]
z=[150,100,374,570,700]


plt.figure(figsize=(15,10));
plt.style.use('seaborn')
le=numpy.arange(len(x))
width=2.5
plt.title("Bar plot")
plt.bar(x,y,color='skyblue',width=5)
plt.bar(x,z,color='yellow',width=5)
#plt.plot(x,y,color='black')
plt.plot(x,z,color='black')
plt.legend(["x","y","z"])
```

## Pie-chart plot

```python
plt.style.use('dark_background')
plt.figure(figsize=(15,10));
slices=[30,20,9,45,34,23,78]
label=['sec1','sec2','sec3','sec4','sec5','sec6','sec7'] ##stuff names as labels-
exp=[0.1,0,0,0,0,0,0.1]
ag=90
tit='GDP plot'
plt.pie(slices,wedgeprops={'edgecolor':'black'},labels=label,explode=exp,startangle=ag)
plt.title('GDP plots')
plt.tight_layout()
plt.legend()
plt.show()
```

## Bar and scatter plot overlap 1-

```python
x=ope['Year']
y=ope['USA']
z=ope['Japan']
plt.style.use('dark_background')
plt.figure(figsize=(15,10))
plt.scatter(x,y,label='USA')
plt.scatter(x,z,label='Japan')
plt.plot(x,y,label='USA',color='blue')
plt.plot(x,z,label='Japan',color='yellow')
plt.bar(x,y,label='USA-Bar',color='skyblue')
plt.legend()
plt.tight_layout()
plt.xticks(x)
plt.savefig('bar plot1',dpi=100)
#plt.yticks(y)
plt.show()
```

## 2-

```python
x=ope['Year']
y=ope['USA']
z=ope['Japan']
wid=0.2
plt.style.use('dark_background')
plt.figure(figsize=(15,10))
plt.scatter(x,y,label='USA')
plt.scatter(x,z,label='Japan')
#plt.plot(x,y,label='USA',color='blue')
#plt.plot(x,z,label='Japan',color='yellow')
plt.bar(x-wid,y,label='USA-Bar',color='skyblue',width=0.45)
plt.bar(x+wid,z,label='Japan',color='orange',width=0.45)
plt.legend()
plt.tight_layout()
plt.xticks(x)
plt.savefig('bar plot',dpi=100)
#plt.yticks(y)
plt.show()
```
